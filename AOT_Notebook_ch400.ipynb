```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
results_df = pd.read_csv('EvaluationData/DimensionData.csv', sep=",")
results_df
```

```python
model_list = ["DE_ML", "PSO_ML", "RS_ML"]
```

```python
for model in model_list:
    de_df = results_df[model]
    mean_df = results_df.mean()
    results_df["DE"] = results_df["DE_ML"] - mean_df["DE_ML"]
    results_df["RS"]  = results_df["RS_ML"] - mean_df["RS_ML"]
    results_df["PSO"] = results_df["PSO_ML"] - mean_df["PSO_ML"]
```

```python
results_df
```

```python
results_data = results_df[["DE", "RS", "PSO"]]
```

```python
PROPS = {'boxprops':{'facecolor': "white", 'edgecolor':'black'},
'medianprops':{'color':'orange'},
'whiskerprops':{'color':'black'},
'capprops':{'color':'black'}
}
sns.boxplot(data=results_data, color="white", linewidth=1.2, flierprops = dict(markerfacecolor='black', markersize=8,
                  markeredgecolor='none'), **PROPS)
    
```
